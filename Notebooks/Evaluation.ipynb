```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model
from tensorflow.keras.models import load_model
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

# Load Model

```python
model1 = load_model('../models/eye-model_cv1_01-32-58_10182019.h5')
model2 = load_model('../models/eye-model_cv2_01-32-45_10182019.h5')
model3 = load_model('../models/eye-model_cv3_01-32-28_10182019.h5')
model4 = load_model('../models/eye-model_cv4_01-32-15_10182019.h5')
model5 = load_model('../models/eye-model_cv5_01-32-02_10182019.h5')

models = [model1, model2, model3, model4, model5]
```

# Plot Model 

```python
for model in range(len(models)): 
    plot_model(models[model], to_file='model_plot{}.png'.format(model), show_shapes=True)
```

# Load Metrics

```python
model_history1 = pd.read_csv("../models/eye-model-history_cv1_01-32-58_10182019.csv")
model_history2 = pd.read_csv("../models/eye-model-history_cv2_01-32-45_10182019.csv")
model_history3 = pd.read_csv("../models/eye-model-history_cv3_01-32-28_10182019.csv")
model_history4 = pd.read_csv("../models/eye-model-history_cv4_01-32-15_10182019.csv")
model_history5 = pd.read_csv("../models/eye-model-history_cv5_01-32-02_10182019.csv")

model_acc1 = model_history1[['epoch', 'acc', 'val_acc']]
model_loss1 = model_history1[['epoch', 'loss', 'val_loss']]

model_acc2 = model_history2[['epoch', 'acc', 'val_acc']]
model_loss2 = model_history2[['epoch', 'loss', 'val_loss']]

model_acc3 = model_history3[['epoch', 'acc', 'val_acc']]
model_loss3 = model_history3[['epoch', 'loss', 'val_loss']]

model_acc4 = model_history4[['epoch', 'acc', 'val_acc']]
model_loss4 = model_history4[['epoch', 'loss', 'val_loss']]

model_acc5 = model_history5[['epoch', 'acc', 'val_acc']]
model_loss5 = model_history5[['epoch', 'loss', 'val_loss']]
```

# Visualize Data 

```python
ax = plt.gca()

model_acc1.plot(kind='line', x='epoch', y='acc', color='green', ax=ax, figsize=(8,8))
model_acc1.plot(kind='line', x='epoch', y='val_acc', color='red', ax=ax)

model_acc2.plot(kind='line', x='epoch', y='acc', color='purple', ax=ax, figsize=(8,8))
model_acc2.plot(kind='line', x='epoch', y='val_acc', color='blue', ax=ax)

model_acc3.plot(kind='line', x='epoch', y='acc', color='yellow', ax=ax, figsize=(8,8))
model_acc3.plot(kind='line', x='epoch', y='val_acc', color='black', ax=ax)

model_acc4.plot(kind='line', x='epoch', y='acc', color='cyan', ax=ax, figsize=(8,8))
model_acc4.plot(kind='line', x='epoch', y='val_acc', color='pink', ax=ax)

model_acc5.plot(kind='line', x='epoch', y='acc', color='orange', ax=ax, figsize=(8,8))
model_acc5.plot(kind='line', x='epoch', y='val_acc', color='brown', ax=ax)

ax.get_legend().remove()
```

```python
ax = plt.gca()

model_loss1.plot(kind='line', x='epoch', y='loss', color='green', ax=ax, figsize=(8,8))
model_loss1.plot(kind='line', x='epoch', y='val_loss', color='red', ax=ax)

model_loss2.plot(kind='line', x='epoch', y='loss', color='purple', ax=ax, figsize=(8,8))
model_loss2.plot(kind='line', x='epoch', y='val_loss', color='blue', ax=ax)

model_loss3.plot(kind='line', x='epoch', y='loss', color='yellow', ax=ax, figsize=(8,8))
model_loss3.plot(kind='line', x='epoch', y='val_loss', color='black', ax=ax)

model_loss4.plot(kind='line', x='epoch', y='loss', color='cyan', ax=ax, figsize=(8,8))
model_loss4.plot(kind='line', x='epoch', y='val_loss', color='pink', ax=ax)

model_loss5.plot(kind='line', x='epoch', y='loss', color='orange', ax=ax, figsize=(8,8))
model_loss5.plot(kind='line', x='epoch', y='val_loss', color='brown', ax=ax)
ax.get_legend().remove()
```
